```python
mysql_user = 'root'
mysql_password = ''
mysql_host = '127.0.0.1'
mysql_db = 'hackreactor_talk'
db_connection = create_engine('mysql+pymysql://' + mysql_user + ':' + mysql_password + '@' + mysql_host + '/' + mysql_db)
```

```python
query = """
            SELECT * 
            FROM actions
            LIMIT 10
        """
pandas.read_sql(query, db_connection)
```

```python
query = """
            SELECT team, COUNT(id) AS impressions
            FROM actions
            WHERE type = "impression"
            GROUP BY team
        """
impressions = pandas.read_sql(query, db_connection)
```

```python
impressions
```

```python
query = """
            SELECT team, COUNT(id) AS clicks
            FROM actions
            WHERE type = "click"
            GROUP BY team
        """
clicks = pandas.read_sql(query, db_connection)
```

```python
clicks
```

```python
imp_clk = pandas.merge(impressions, clicks, how='left', on='team')
imp_clk['click through rate'] = imp_clk['clicks'] / imp_clk['impressions']
imp_clk.sort_values(by='click through rate', ascending=False)
```
